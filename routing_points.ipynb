```python
import requests
```

```python
HASH_TOKEN = 'bmEzY3J1eGF1YnxwZGpJTUdPQXZ4N3VPVW93bjZwRk8ydHdRT1dNZjU2SDlFTlB1UkhU'
APP_ID = 'na3cruxaub'
TOKEN_URL = 'https://api.iq.inrix.com/auth/v1/appToken'

def get_token():
    #Pass in the app_id and hash_token as query parameters
    params = {'appId': APP_ID,
              'hashToken': HASH_TOKEN}
    response = requests.get(TOKEN_URL, params=params)
    response.raise_for_status()  # Raise HTTPError for bad responses
    data = response.json()
    token = data['result']['token']
    return token, response.status_code
```

```python
def get_route(wp_1, wp_2, departure_time):
    '''
    wp_1 and wp_2 are str(lat, long) points.
    
    departure_time: default: now, DepartureTime must be within nine months from the time of the API call, and must occur in the future., 
     format:YYYY-MM-DDTHH:MM:SSZ; for example 2009-04-04T13:42:41Z.
    '''
    token, status = get_token()
    auth_token = 'Bearer '+token
    findRoute_url = 'https://api.iq.inrix.com/findRoute'
    
    # wp_1 = '37.80300372563528, -122.42538202577136'
    # wp_2 = '37.76799081681866, -122.41476498906668'
    # departure_time = '2023-12-06T16:42:41Z' 
    
    # useTraffic = 'true'
    # routeType = '0'
    
    format='json'
    routeOutputFields = 'P,summary'
    
    route = requests.get(findRoute_url, headers={'Authorization':auth_token}, 
                         params={'wp_1':wp_1, 
                                 'wp_2':wp_2,
                                 # 'routeType':routeType,
                                 # 'useTraffic':useTraffic,
                                 'format':format,
                                 'routeOutputFields':routeOutputFields,
                                 'departure_time':departure_time})
    data = route.json()
    result = data['result']['trip']
    
    # routes=[]
    # for i in range(len(data['result']['trip']['routes'])):
        # route = data['result']['trip']['routes'][i]
    route = data['result']['trip']['routes'][0]
    out={'travelTimeMinutes':route['travelTimeMinutes'],
       # 'uncongestedTravelTimeMinutes':route['uncongestedTravelTimeMinutes'],
       'averageSpeed':route['averageSpeed'], 
       'totalDistance':route['totalDistance'],
       'points': route['points']['coordinates']}
        # routes.append(d)

    # output = {'tripId':result['tripId'], 'routes':routes}    
    # return data, output
    return out
    
```
